```python
import requests
import pandas as pd


params = {
    'CountryCode': 'US',
    'subjectId': '13',
}

response = requests.get('https://www.ebooks.com/api/search/subject/', params=params)

```

```python
response.json().keys()
```

```python
response.json()['books']
```

```python
result_json=response.json()['books']
```

```python
#title
result_json[0]['title']
```

```python
# subtitle
result_json[0]['subtitle']
```

```python
#publisher
result_json[0]['publisher']
```

```python
# publication_year
result_json[0]['publication_year']
```

```python
#price
result_json[0]['price']
```

```python
#author
result_json[0]['authors'][0]['name']
```

```python
title=[]
subtitle=[]
author=[]
publisher=[]
publication_year=[]
price=[]
for result in result_json:
    title.append(result['title'])
    subtitle.append(result['subtitle'])
    author.append(result['authors'][0]['name'])
    publisher.append(result['publisher'])
    publication_year.append(result['publication_year'])
    price.append(result['price'])
```

```python
my_df=pd.DataFrame({'Author':author,'Title':title,'Subtitle':subtitle,'Publisher':publisher,'Publication_year':publication_year,
    
'Price':price})
```
